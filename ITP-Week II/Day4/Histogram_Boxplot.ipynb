# Histogram

- Frequency vs Range
- A histogram is a graphical display of data using bars of different heights. 
- In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range.
- A histogram displays the shape and spread of continuous sample data. 

- Only one axis is required

## Using df.plot 

```python
df.plot(kind = "hist" , y = "SalePrice")
plt.show()
```

```python
df.plot(kind = "hist" , y = "SalePrice" , bins =25 , edgecolor = "white" , legend = False )
plt.show()
```
