```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#To fit the graph within resolution
%matplotlib inline
```

#### Line plot 
- A line graph is commonly used to display change over time as a series of data points connected by straight line segments on two axes. The line graph therefore helps to determine the relationship between two sets of values, with one data set always being dependent on the other set.

```python
x = np.linspace (1,9,9)
y= x**x
print(x)
print(y)
```

```python
plt.plot(x,y)  #It will show you the object and graph
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
plt.plot ([1,3,2,4,5,6,3], [1,4,16,25,36,17,12])  # ==> Order of the Value MATTERS
plt.show()
```

### Bar Graph - Using pandas.Series.plot

```python
x = pd.Series ([100,300,250,500,400], index=['a','b','c','d','e'])
x.plot( kind = 'bar' )
```

#### This x. plot function belongs to Pandas Package and not matlib

Will have CATEGORICAL values in x-axis

When to use: If you have comparative data that you would like to represent through a chart then a bar chart would be the best option. This type of chart is one of the more familiar options as it is easy to interpret. These charts are useful for displaying data that is classified into nominal or odinal categories.

### Pie Chart

```python
x = pd.Series ([100,300,250,500,400], index=['a','b','c','d','e'])
x.plot( kind = 'pie' )
plt.show()
```

### Area Plot

```python
x = pd.Series ([100,300,250,500,400], index=['a','b','c','d','e'])
x.plot( kind = 'area' )
plt.show()
```

### Box Plot

```python
x = pd.Series ([100,300,250,500,400], index=['a','b','c','d','e'])
x.plot( kind = 'box' )
plt.show()
```
