#### map()

```python
import pandas as pd
```

```python
df = pd.DataFrame ( {"Fruit" : [ "Apple","Grapes","Apple","Grapes","Gauva", "Gauva","Berry","Berry","Mango"],
"Color" : ["Red","Green","Green", "Black","Green","Red","Black","Red","Yellow"],
"Cost" : [50,70,80,50,65,70,80,90,100]} )
df
```

```python
s = {"Apple": "Good", "Grapes": "Better","Gauva": "Best","Berry":"Worst"}
```

```python
df['Quality'] = df['Fruit'].map( s )
df
```

### apply()

```python
def discount(a):
    return a-(a*0.05)

df[ "Discounted_Cost"]=df["Cost"].apply(discount)
df
```
