```python
import pandas as pd
Table1 = pd.DataFrame({'A': ['10', '11', '12', '13'],
                    'B': ['20', '21', '22', '23'],
                    'C': ['30', '31', '32', '33'],
                    'D': ['40', '41', '42', '43']},
                    index=[0, 1, 2, 3])
Table2 = pd.DataFrame({'A': ['01', '02', '03', '04'],
                    'B': ['24', '25', '26', '27'],
                    'C': ['C1', 'C2', 'C3', 'C4'],
                    'D': ['44', '45', '46', '47']},
                    index=[4, 5, 6, 7])
Table3 = pd.DataFrame({'A': ['18', '19'],
                    'B': ['28', '29'],
                    'C': ['38', '39'],
                    'D': ['48', '49']},
                   index=[8, 9])
df = pd.concat([Table1,Table2,Table3], axis = 0)
df
```

#### With 1 different column

```python
import pandas as pd
Table1 = pd.DataFrame({'A': ['10', '11', '12', '13'],
                    'B': ['20', '21', '22', '23'],
                    'C': ['30', '31', '32', '33'],
                    'D': ['40', '41', '42', '43']},
                    index=[0, 1, 2, 3])
Table2 = pd.DataFrame({'A': ['01', '02', '03', '04'],
                    'K': ['24', '25', '26', '27'],
                    'B': ['24', '25', '26', '27'],
                    'C': ['C1', 'C2', 'C3', 'C4'],
                    'D': ['44', '45', '46', '47']},
                    index=[4, 5, 6, 7])
df = pd.concat([Table1,Table2,Table3], axis = 0,sort = True)
df
```

#### All Different Columns

```python
import pandas as pd
Table1 = pd.DataFrame({'L': ['10', '11', '12', '13'],
                    'C': ['20', '21', '22', '23'],
                    'B': ['30', '31', '32', '33'],
                    'D': ['40', '41', '42', '43']},
                    index=[0, 1, 2, 3])
Table2 = pd.DataFrame({'I': ['01', '02', '03', '04'],
                    'K': ['24', '25', '26', '27'],
                    'J': ['C1', 'C2', 'C3', 'C4'],
                    'A': ['44', '45', '46', '47']},
                    index=[4, 5, 6, 7])
df=pd.concat( [Table1,Table2] , axis =0, sort=True)
df
```
