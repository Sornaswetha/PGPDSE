```python
import pandas as pd
```

```python
left_frame = pd.DataFrame({'key': range(5), 
                           'left_value': ['a', 'b', 'c', 'd', 'e']})
left_frame
```

```python
right_frame = pd.DataFrame({'key': range(2, 7), 
                           'right_value': ['f', 'g', 'h', 'i', 'j']})
right_frame
```

```python
df_inner = pd.merge(left_frame, right_frame, on = 'key', how= "inner")
df_inner
```

```python
df_outer = pd.merge(left_frame, right_frame, on = 'key', how= "outer")
df_outer
```

```python
df_left = pd.merge(left_frame, right_frame, on = 'key', how= "left")
df_left
```

```python
df_right = pd.merge(left_frame, right_frame, on = 'key', how= "right")
df_right
```

```python
df_right = pd.merge(right_frame,left_frame, on = 'key', how= "right")
df_right
```
