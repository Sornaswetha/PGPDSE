### Rank()

```python
df['Rank'] = df["Math"].rank()
df  #Average of same enteries
```

```python
df['Rank'] = df["Math"].rank( method = 'average')  # Default
df  #Average of same enteries
```

```python
df['Rank'] = df["Math"].rank( method = 'min')
df  #Average of same enteries
```

```python
df['Rank'] = df["Math"].rank( method = 'dense')
df  #Average of same enteries
```

```python
df['Rank'] = df["Chem"].rank(ascending = False)
df  #Average of same enteries
```
