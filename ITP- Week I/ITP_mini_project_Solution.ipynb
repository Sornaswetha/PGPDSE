```python
import math

# To check if the given input is a string of float valus
def is_float (f1,f2=1):
    try:
        float (f1) and float (f2)
    except ValueError: 
        return 0
    else:  
        return 1

# Area 
def area_square (side):
    return float(side)* float(side)

def area_rectangle (length, breadth):
    return float(length) *float (breadth) 

def area_triangle (base,height):
    return (float(base) * float (height)) /2 

def area_circle (radius):
    return  math.pi * (float(radius)**2) 

# Function for selecting 
def fun(choice):
 # Square
    if  choice == 1 :  
        print ("Square", )
        side =  input ( "Please enter the length of a side: ")
        if is_float (side):
            print( "Area of Square is ", area_square(side) )
        else:
            side= float ( input ("Please enter a valid length"))
            print( "Area of Square is ", area_square(side) )
            
#Rectange
    elif choice == 2 :  
        length = input ("Please enter valid length : ")
        breadth = input ("Please enter valid breadth : ")
        if is_float (length,breadth):
            print ( "Area of Rectngle is" , area_rectangle (length,breadth))
        else:
            length = float (input ("Please enter valid length : "))
            breadth = float (input ("Please enter valid breadth : ") )
            print ( "Area of Rectngle is" , area_rectangle (length,breadth))

#Triangle
    elif choice == 3:
        base =  input("Please enter the base : ")
        height = input ("Please enter the height : ")
        if is_float(base,height):
            print ("Area of Triangle is" , area_triangle (base,height))
        else:
            base = float (input("Please enter valid base : "))
            height = float (input ("Please enter valid height : "))
            print ("Area of Triangle is" , area_triangle (base,height))
#Circle
    elif choice == 4:
        radius =input ( "Please enter the radius : ")
        if is_float(radius):
            print ("Area of Circle is %0.2f" % area_circle(radius) )
        else:
            radius = float (input ( "Please enter valid radius : "))
            print ("Area of Circle is %0.2f" % area_circle(radius) )
#To check for any other options.
#To check if it is string or a number greater than 4
    else:
        a= input ("Invalid input, Please enter a vaild option !!! ")
        try:
            int(a)
        except ValueError:
            a= input ("Invalid input, Please enter a vaild option !!! ")
            choice = int (a)
            fun(choice)   
        else:
            choice = int (a)
            fun(choice)    
            
#MAIN

print ( """Which shape would you like to calculate the area for? Please enter the option number-
1. Square
2. Rectangle
3. Triangle
4. Circle""" )
a =  input ("Enter Option: ")
# To Check if we are getting an integer (1-4) or any other string Value
try:
    int(a)
except ValueError: # If a string , get the input and run the function
    a= input ("Invalid input, Please enter a vaild option !!! ")
    choice = int (a)
    fun(choice)   
else:  # If input is between 1-4 call the function
    choice = int (a)
    fun(choice)    
```

### Problem 2:

Create and encrypter in python based on the ceaser cipher. It is a substitution cipher where each character of the original text is shifted a certain number characters in the alphabet. Write a function that would require 2 arguments – the input text to be encrypted and a key. For eg: Given the input text ‘hello’ and the key 3, the resulting encryted text would be ‘khoor’. Here you can see that every character in the string hello is shifted by 3 characters. ‘h’ has shifted to ‘k’, ‘e’ has shifted to ‘h’ and so on. If a key of 5 were used, the resulting string would be ‘mjqqt’. This function should be capable of ignoring any characters which are not alphabets. Th2 character ‘z’ entered b
y the user for a key of 3 would result in ‘c’.

Usage:<br>
**encrypt(‘hello world!’, 3)**<br>
**‘khoor zruog!’**

Similarly create decrypter which can decode the encryted text when provided the input text and key

Usage:<br>
**decrypt(‘khoor zruog!’, 3)**<br>
**‘hello world!’**

For the sake of simplicity you can assume that input solely consists of lowercase alphabets, spaces and punctuation symbols. Numbers in the input text would also be ignored similar to symbols.

```python
# E =(x+n)\mod {26}.
# D =(x-n)\mod {26}.
# To get the ascii Value +/- 97

def fun_encrypt(string ,n):
    encrypt = ""
    for char in string:
        if ord(char) >= ord('a')   and ord (char) <= ord ('z'):
            encrypt += chr((ord(char) + n - 97) % 26 + 97)
        else:
            encrypt +=char
    return encrypt

def fun_decrypt(string , n ):
    decrypt = ""
    for char in string:
        if ord(char) >= ord('a')   and ord (char) <= ord ('z'):
            decrypt += chr((ord(char) - n - 97) % 26 + 97)
        else:
            decrypt +=char
    return decrypt

# Main 

string = input ("Enter a string: ")
n = int ( input ("Enter key : "))

print (" Encrypted Value : ",  fun_encrypt (string ,n) )
print (" Decrypting the encrypted Value: ", fun_decrypt (fun_encrypt (string ,n)  , n ))
```
